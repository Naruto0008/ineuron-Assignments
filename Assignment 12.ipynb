```python
import PyPDF2
f=open('Jigs and Fixtures Design Manual (Prakash Joshi) (z-lib.org).pdf','rb')
f1=open('Handbook of Die Design (Ivana Suchy) (z-lib.org).pdf','rb')
pdfreader=PyPDF2.PdfFileReader(f)
pdfreader1=PyPDF2.PdfFileReader(f1)
pdfwriter=PyPDF2.PdfFileWriter()

for page in range(pdfreader.numPages):
    page=pdfreader.getPage(page)
    pdfwriter.addPage(page)

for page in range(pdfreader1.numPages):
    page=pdfreader1.getPage(page)
    pdfwriter.addPage(page)

pf=open('combined.pdf','wb')
pdfwriter.write(pf)
pf.close()
f.close()
f1.close()
```

```python
rb,wb are the modes to be used for PdfFileReader(),PdfFileWriter()
```
